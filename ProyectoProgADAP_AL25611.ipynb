 4. ¿Que porcentaje de personas sufren las condiciones de riesgo en el estudio:
 Hipertensión, Diabetes,Tabaquismo, Historial Familiar, Consumo de Alcohol, Angina inducida por ejercicio?

 -En este caso se concentro en los factores de riesgo y cueantas personas lo padecen, la mayoria son respuestas binarias como 1 o 0 o y o n. Estos datos categoricos se consultaron para ver cuales de ellos son los que mas lo presentaban en el estudio. El consumo de alcohol fue lo que prevalecio en esta categoria.

```python

import pandas as pd
import matplotlib.pyplot as plt

# Cargar archivo
file = "/content/drive/MyDrive/ClassFiles/heart_attack_dataset-Clean.csv"
df = pd.read_csv(file)


# funcion porcentajes
def porcentaje(columna, tipo='binario'):
    if tipo == 'binario':
        return round((df[columna].sum() / len(df)) * 100, 2)
    elif tipo == 'y/n':
        return round((df[columna].str.lower().value_counts().get('y', 0) / len(df)) * 100, 2)
    elif tipo == 'mayor_que_0':
        return round((df[df[columna] > 0].shape[0] / len(df)) * 100, 2)

# Diccionario de condiciones
condiciones = {
    'Hipertension': ('hypertension', 'binario'),
    'Diabetes': ('diabetes', 'binario'),
    'Tabaquismo': ('smoker', 'binario'),
    'Historial Familiar': ('familyhistory', 'binario'),
    'Consumo de Alcohol': ('alcoholconsumption', 'mayor_que_0'),
    'Angina inducida por ejercicio': ('exerciseinducedangina', 'y/n')
}

#
porcentajes = {
    nombre: porcentaje(col, tipo)
    for nombre, (col, tipo) in condiciones.items()
}

# resultados
porciento = pd.DataFrame.from_dict(porcentajes, orient='index', columns=['% de personas'])
print(porciento)
print("***" * 40 + "\n")
print("***" * 40 + "\n")

# Grafica
plt.figure(figsize=(10, 6))
plt.plot(
    porciento.index,
    porciento['% de personas'],
    marker='o',
    linestyle='-',
    color='red'
)
plt.title('Porcentaje de Personas con Condiciones de Riesgo')
plt.ylabel('Porcentaje (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

2.  Con correlaciones de datos realmente numericos y con el outcome (infarto cierto), ¿Que correlacion de variables hay?

El heatmap nos indica definitivamente que las variables no tienen una correlacion entre ellas, lo que puede ser indicativo que una sola variable no tiene influencia sobre si hubo un evento de ataque cardiaco.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Cargar archivo
file = "/content/drive/MyDrive/ClassFiles/heart_attack_dataset-Clean.csv"
df = pd.read_csv(file)

# Agregar columna outcome_fix
df['outcome_fix'] = df['outcome'].str.lower().apply(lambda x: 1 if 'heart attack' in x else 0)

# columnas numericas
num_col = df[['age', 'cholesterol', 'bloodpressure', 'heartrate', 'bmi',
                    'income', 'maxheartrate', 'st_depression']]

# Filtrar solo registros con ataque cardiaco cierto
df_heart_attack = df[df['outcome_fix'] == 1]

# Agregar 'outcome_encoded' al subset filtrado
subset_continuas = df_heart_attack[num_col.columns].copy()
subset_continuas['outcome'] = df_heart_attack['outcome_fix']

#  matriz de correlacion con datos de hearattack
heartattack = subset_continuas.corr()

# GRafica
plt.figure(figsize=(13, 8))
sns.heatmap(heartattack, annot=True, fmt=".2f", cmap="Greens", linewidths=0.5)
plt.title('Correlación * Casos de infarto ciertos')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

3. ¿Cuales son las media de variables age,cholesterol, bloodpressure y heartrate con datos de ataque cardiaco cierto y sin ataque?

-Se tomaron estas variables para ponerlas a prueba que tanto es la influencia entre una ataque cardiaco o no. Podemos comprobar como en el punto anterior que no son definitivas para predecir un ataque cardiaco, los resultados son muy similares, muy poca variacion, en provocar el evento.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar archivo
file = "/content/drive/MyDrive/ClassFiles/heart_attack_dataset-Clean.csv"
df = pd.read_csv(file)

# columna outcome  0 y 1
df['outcome_fix'] = df['outcome'].map({'Heart Attack': 1, 'No Heart Attack': 0})

# medias por grupo
variables = ['age', 'cholesterol', 'bloodpressure', 'heartrate']
medias_por_grupo = df.groupby('outcome_fix')[variables].mean().T

# Renombrar columnas para que sean legibles
medias_por_grupo.columns = ['Sin Ataque', 'Con Ataque']

# Graficar
colores = ['mediumseagreen', 'red']
medias_por_grupo.plot(kind='bar', figsize=(10, 7), color=colores)
plt.title('Medias de Variables Ataque Cardiaco')
plt.ylabel('Media')
plt.xlabel('Variable')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Grupo')
plt.tight_layout()
plt.show()
```

4. ¿Que grupos etnicos tuvieron mayores ataques cardiacos?

-Vemos en esta parte que tuvo mayores eventos de ataque cardiacos fueron de Asia. Por el contrario , el grupo white fue el menos afectado. Podemos observar una variable que sí podria estar relacionada con un evento de estas caracteristicas.

```python
import pandas as pd
import matplotlib.pyplot as plt


# Cargar archivo
file = "/content/drive/MyDrive/ClassFiles/heart_attack_dataset-Clean.csv"
df = pd.read_csv(file)

# columna outcome  0 y 1
df['outcome_fix'] = df['outcome'].map({'Heart Attack': 1, 'No Heart Attack': 0})

# Agrupar por etnia y outcome
cont = df.groupby(['ethnicity', 'outcome_fix']).size().unstack(fill_value=0)
cont.columns = ['No Ataque', 'Ataque']
cont = cont.sort_index()

# Grafico
plt.figure(figsize=(10, 6))
plt.plot(cont.index, cont['No Ataque'], marker='o', linestyle='-', label='No Ataque', color='skyblue')
plt.plot(cont.index, cont['Ataque'], marker='o', linestyle='-', label='Ataque', color='salmon')
plt.title('Ataques Cardiacos por Grupo Etnico')
plt.xlabel('Grupo Etnico')
plt.ylabel('Cantidad de Personas')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Condición')
plt.tight_layout()
plt.show()
```
