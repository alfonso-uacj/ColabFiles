2. Crear los siguientes gráficos con Seaborn:

a) Mapa de Calor (Heatmap):



*   Objetivo: Visualizar la correlación entre las variables numéricas en el dataset.
*   Instrucciones: Utiliza un mapa de calor para mostrar la matriz de correlación entre las variables numéricas (total_bill, tip, size).
*   Asegúrate de personalizar el mapa con una paleta de colores y etiquetas en los ejes.



```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')

# Asignar a corr la correlacion .corr del dataset
corr = df[['total_bill', 'tip', 'size']].corr()

# crear el heatmap (plt contiene) sns formatea
plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, cmap='Spectral', fmt=".3f", linewidths=1.5, annot_kws={"size": 8, "color": "black"})

# etiquetas
plt.title("Heatmap correlacion TIP")
plt.xlabel("Variables eje x", fontweight="bold")
plt.ylabel("Variables eje y", fontweight="bold")

# mostrar
plt.show()
```

b) Diagrama de Violín:



*   Objetivo: Mostrar la distribución de las propinas (tip) para diferentes días de la semana.
*   Instrucciones: Utiliza un diagrama de violín para comparar cómo varía la distribución de las propinas en cada uno de los días (day).
*   Personaliza el gráfico con colores y etiquetas adecuadas.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset from Seaborn
df = sns.load_dataset('tips')

#formato para backgound diferente
sns.set_style("whitegrid")

# crear el diagrama de violin 'tip' por 'day'
plt.figure(figsize=(9, 6))
#formato del diagrama
sns.violinplot(data=df, hue='day', y='tip', palette='coolwarm', inner="box", linewidth=2)


# Etiquetas
plt.xlabel("Dias", color="red")
plt.ylabel("Tips", color="blue")
plt.title("Distribucion Tips * Dias", fontweight="bold")

# Mostrar
plt.show()
```

c) Gráfico de dispersión (Scatter Plot):



*   Objetivo: Analizar la relación entre el total de la cuenta (total_bill) y la propina (tip).
*   Instrucciones: Crea un gráfico de dispersión con total_bill en el eje X y tip en el eje Y. Usa un color diferente para cada día de la semana.



```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')

# Diseñar un grafico de dispercion con un color diferente por dia de la semana
colores_dias = {"Thur": "green", "Fri": "red", "Sat": "orange", "Sun": "purple"}
plt.figure(figsize=(12, 5))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day', palette=colores_dias, edgecolor='blue')

# Etiquetas
plt.xlabel("Total Bill ($)", fontsize=14, color="blue")
plt.ylabel("Tip Amount ($)", fontsize=14, color="red")
plt.title("Relación entre el total de la cuenta (total_bill) y la propina (tip)", fontsize=15, fontweight="bold", color="orange")

# Mostrar grafica
plt.show()
```

d) Gráfico de barras (Bar Plot):



*   Objetivo: Comparar el promedio de la propina (tip) por género (sex).
*  Instrucciones: Utiliza un gráfico de barras para mostrar el promedio de las propinas por género (hombres y mujeres).


```python
import seaborn as sns


df = sns.load_dataset('tips')

# Calcular desde datos del dataset las propinas, ( en este caso en un inicio marcaba un error por no poner observed=True )
tips_genero = df.groupby('sex', observed=True)['tip'].mean().reset_index()

# Crear la grafica de barras y formato
plt.figure(figsize=(7, 5))
sns.barplot(data=tips_genero, hue='sex', y='tip', palette='Blues', width=0.3, edgecolor="blue")

# Etiqueta
plt.xlabel("Genero", fontsize=10, color="red")
plt.ylabel("Tip promedio", fontsize=10, color="green")
plt.title("Tip promedio por genero", fontsize=16, fontweight="bold", color="brown")

# Mostrar grafica
plt.show()
```
