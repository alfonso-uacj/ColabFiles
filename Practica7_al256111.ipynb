a) Gráfico de barras:

Objetivo: Comparar el promedio de largo y ancho de los pétalos entre las tres especies de Iris.
Instrucciones: Utiliza un gráfico de barras para mostrar la media de las columnas petal_length y petal_width agrupadas por species.

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('iris') # se carga dataset

# media de las columnas petal_length y petal_width agrupadas por species.
media = df.groupby('species')[['petal_length', 'petal_width']].mean()

# Crear gráfico de barras una vez que se calculo los datos
#media.plot(kind='bar', figsize=(5, 4)) #ete fue con opcion "bar"
#media.plot(kind='line', figsize=(5, 4)) #este es con opcion line
media.plot(kind='barh', figsize=(5, 4)) #este es con opcion line
# Configurar etiquetas y título
plt.xlabel("Especie")
plt.ylabel("Media")
plt.title("Media -Petal Length y Petal Width por Especie")
plt.legend(title="Valores")

# Mostrar gráfico
plt.show()

```

b) Histograma:

Objetivo: Visualizar la distribución de los largos de los sépalos (sepal_length).
Instrucciones: Crea un histograma de la variable sepal_length y asegúrate de que tenga un número adecuado de bins. Coloca etiquetas en los ejes y un título.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

# Crear el histograma de 'sepal_length'
plt.figure(figsize=(9, 6))
#plt.hist(df['sepal_length'], bins=14, edgecolor='yellow', alpha=0.7) #color default sin indicar
plt.hist(df['sepal_length'], bins=14, edgecolor='yellow', color= 'brown', alpha=0.7)
plt.xlabel("Longitud de sepal")
plt.ylabel("Frecuencia")
plt.title(" histograma de la variable sepal_length")

# Mostrar gráfico
plt.show()
```

c) Gráfico de dispersión (scatter plot):

Objetivo: Relacionar el largo del pétalo con el ancho del pétalo.
Instrucciones: Crea un gráfico de dispersión que compare petal_length con petal_width. Usa un color diferente para cada especie.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

# Crear gráfico de dispersión
#opciones para scatterplot diagrama x, y titulos; hue, dependiendo de la etiqueta colorea (species); palette es para una configuracion preestablecida (colores)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', palette='Set2', edgecolor='black')

# Etiquetas
plt.xlabel("Longitud del pétalo (cm)")
plt.ylabel("Ancho del pétalo (cm)")
plt.title("Gráfico de Dispersión de Longitud vs Ancho del Pétalo")

# Mostrar el gráfico
plt.show()




```

d) Gráfico de cajas (box plot):

Objetivo: Analizar la distribución de las medidas del sépalo (sepal_length, sepal_width).
Instrucciones: Crea un gráfico de cajas para mostrar la distribución de las variables sepal_length y sepal_width por especie.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

# gráfico de cajas 'sepal_length' y 'sepal_width' por especie
fig, axes = plt.subplots(1, 2, figsize=(9, 4))

# Gráfico de cajas para sepal_length
sns.boxplot(data=df, y='sepal_length',  hue='species', palette='Set1',  ax=axes[0])
axes[0].set_title(" Sepal Length por Especie")
axes[0].set_xlabel("Especie")
axes[0].set_ylabel("Sepal Length")

# Gráfico de cajas para sepal_width
sns.boxplot(data=df, y='sepal_width',  hue='species', palette='Set2', ax=axes[1])
axes[1].set_title("Sepal Width por Especie")
axes[1].set_xlabel("Especie")
axes[1].set_ylabel("Sepal Width")

# Mostrar el gráfico
plt.show()
```

e) Gráfico de líneas:

Objetivo: Mostrar la tendencia de la longitud del pétalo (petal_length) a través de las observaciones.
Instrucciones: Crea un gráfico de líneas utilizando la variable petal_length (en el eje Y) con el índice de las observaciones (en el eje X). Puedes hacer esto para ver cómo varía la longitud del pétalo a lo largo del dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset Iris desde Seaborn
df = sns.load_dataset('iris')

# Crear gráfico de líneas de longitud del pétalo
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['petal_length'], marker='*', linestyle='--', color='g', alpha=0.9)
# Etiquetas formato
plt.xlabel("Indice de Observacion")
plt.ylabel("Longitud de Petalo")
plt.title("Variacion de la Longitud del Petalo")
# Mostrar
plt.show()
```
